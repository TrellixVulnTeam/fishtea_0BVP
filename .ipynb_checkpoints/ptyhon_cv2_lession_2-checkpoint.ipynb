```python
import cv2
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt#Matplotlibæ˜¯RGB
%matplotlib inline 

img =  cv2.imread("lena.jpg")

def cv_show(name,img):
    cv2.imshow(name,img) 
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
top_size,bottom_size,left_size,right_size = (80,80,80,80)

replicate = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_REFLECT)
reflect_101 = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_WRAP)
constant = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,cv2.BORDER_CONSTANT,value=0)
title = ["img","replicate","reflect","reflect_101","wrap","constant"]
images = [img,replicate,reflect,reflect_101,wrap,constant]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(title[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img = cv2.imread("lena.png")
cv_show('imng',img)
```

```python
blur = cv2.blur(img,(3,3),)
cv_show('blur',blur)
```

```python
box = cv2.boxFilter(img,-1,(3,3),normalize=True)
cv_show('box',box)
```

```python
box = cv2.boxFilter(img,-1,(3,3),normalize=False)
cv_show('box',box)
```

```python
gs = cv2.GaussianBlur(img,(3,3),1)
cv_show('gs',gs)
```

```python
mean =  cv2.medianBlur(img,3)
cv_show('mean',mean)
```
